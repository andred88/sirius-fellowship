```python
import pandas as pd
import requests 
import json
import numpy as np
```

```python
url = f'https://fakerapi.it/api/v1/persons?_locale=pt_BR&_quantity=300'
```

```python
result = requests.get(url, verify=True).json()

```

```python
df_pessoas = pd.DataFrame(result['data'])
```

```python
df_pessoas['gender'].replace(['male'], 'M', inplace=True)
df_pessoas['gender'].replace(['female'], 'F', inplace=True)
```
